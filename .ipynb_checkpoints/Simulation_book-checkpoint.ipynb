```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcol
import matplotlib.cm as cm

x = y = np.arange(-10,10,0.05)
x1,y1 = np.meshgrid(x,y)

z = (x1**2) - (y1**2)

# Create red-blue colourmap
cm1 = mcol.LinearSegmentedColormap.from_list("MyCmapName",["b","w","r"])

plt.contourf(x,y,z,50,cmap=cm1,vmin=-100,vmax=100)
plt.colorbar()
plt.show()
```

```python
with open('diagnostics.dat') as f_in3:
        t,un,us,pn,ps = np.genfromtxt(f_in3,unpack=True,usecols=[0,1,2,3,4])
f_in3.close()

bCorrect = -250

un = un*np.abs(bCorrect)
us = us*np.abs(bCorrect)
pn = pn*bCorrect
ps = ps*bCorrect

f = un+us

plt.plot(t,f)

plt.plot(t,pn)
plt.plot(t,ps)
plt.show()
```
